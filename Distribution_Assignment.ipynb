3. A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and 55 if the data follows a normal distribution?

Answer - The average (mean) value is 50.

Most data values are close to the average.

The standard deviation is 5, which tells us how far values usually spread from the mean.

Now:

45 is 5 less than 50

55 is 5 more than 50

So, 45 to 55 means one standard deviation below and above the mean.

In a normal (bell-shaped) distribution, about 68% of the values fall within one standard deviation of the mean.

5. What is Kurtosis and their type?

Answer - Kurtosis is a statistical measure that describes the shape of a distribution, especially the peakedness and heaviness of the tails compared to a normal distribution.

Types

1. Mesokurtic ‚Üí Normal shape

2. Leptokurtic ‚Üí Sharp peak, heavy tails

3. Platykurtic ‚Üí Flat peak, light tails

6.Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

Answer - The uniform distribution is a good model for rolling a fair die because each outcome (1 to 6) has the same chance of occurring. Since no number is favored over another, all outcomes are equally likely.

This equal probability of all possible outcomes is exactly the main feature of a uniform distribution, where every value in the range has the same probability.

8. How does log transformation help in stabilizing variance and making data more normally distributed?

Answer - A log transformation is commonly used when data is skewed or shows unequal variance. It helps in two main ways: variance stabilization and normality improvement.

1. Stabilizing Variance (Reducing Heteroscedasticity)
Problem:

In many datasets:

Small values ‚Üí small spread

Large values ‚Üí large spread

So variance increases with magnitude (fan/funnel shape in scatter plot).

log does:

Log compresses large values more than small values.

2. Making Data More Normal (Reducing Skewness)
Problem:

Right-skewed data (income, sales, population, etc.)

Many small values

Few very large values (long tail)

log does:

Pulls long right tail inward

Brings extreme values closer

Makes distribution more symmetric
